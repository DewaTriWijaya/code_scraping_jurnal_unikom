## Cleaning Data

```python
# Tahun publikasi & indeks
crossref['indexed_year'] = pd.to_datetime(crossref['indexed_date_time'], errors='coerce').dt.year
crossref['published_year'] = pd.to_datetime(crossref['published'], errors='coerce').dt.year


# Jumlah penulis
crossref['author_count'] = crossref['authors'].fillna('').apply(lambda x: len(x.split(';')) if x else 0)


# Filter data valid
crossref = crossref[crossref['indexed_year'].notna()]
```

```python
#total data sebelum dihapus duplikat
dataBefore = len(crossref)

# hapusdata duplikat berdasarkan title
crossref = crossref.drop_duplicates(subset=['title'])

#total data setelah dihapus duplikat
dataAfter = len(crossref)

print(f"Total data sebelum dihapus duplikat: {dataBefore}")
print(f"Total data setelah dihapus duplikat: {dataAfter}")
```

```python
# ubah data crossref column author_query menjadi id_author dari OpenAlex berdasarkan pencocokan nama penulis
crossref = crossref.merge(openalex[['nama', 'id_author']], left_on='author_query', right_on='nama', how='left')
```

```python
crossref.head()
```

```python
openalex.head()
```

```python
authors = pd.read_csv('author/openalex/authors_cleaned.csv')

# 1. Pastikan kolom 'fullname' dan 'programs_name' tidak ada di crossref 
# jika tujuannya ingin mengambil data terbaru dari authors_cleaned
cols_to_use = ['fullname', 'programs_name', 'nidn']
authors_subset = authors[cols_to_use].drop_duplicates(subset=['fullname'])

# Menghapus kolom yang sama di crossref (jika ada) sebelum merge untuk menghindari suffix error
if 'programs_name' in crossref.columns:
    crossref = crossref.drop(columns=['programs_name'])
if 'fullname' in crossref.columns:
    crossref = crossref.drop(columns=['fullname'])
if 'nidn' in crossref.columns:
    crossref = crossref.drop(columns=['nidn'])

# 2. Lakukan Merge
crossref = crossref.merge(
    authors_subset,
    left_on='author_query',
    right_on='fullname',
    how='left'
)
```

## 1️⃣ Top 10 Dosen – Publikasi Terbanyak (1 Tahun)

```python
## 1️⃣ Top 10 Dosen – Publikasi Terbanyak (1 Tahun)
year_1 = crossref['indexed_year'].max()


k1 = (crossref[crossref['indexed_year'] == year_1]
.groupby('author_query')['doi']
.count()
.sort_values(ascending=False)
.head(10))


sns.barplot(x=k1.values, y=k1.index)
plt.title('Top 10 Dosen dengan Publikasi Terbanyak (1 Tahun)')
plt.xlabel('Jumlah Publikasi')
plt.ylabel('Dosen')
plt.show()
```

```python
# Tentukan tahun terbaru
year_1 = crossref['indexed_year'].max()

# Cari Top 10 dosen berdasarkan jumlah publikasi
top_10_dosen = (
    crossref[crossref['indexed_year'] == year_1]
    .groupby('author_query')['doi']
    .count()
    .sort_values(ascending=False)
    .head(10)
    .index
)

# Ambil detail publikasi untuk Top 10 dosen tersebut
output_text = (
    crossref[
        (crossref['indexed_year'] == year_1) &
        (crossref['author_query'].isin(top_10_dosen))
    ][[
        'id_author',
        'author_query',
        'programs_name',
        'nidn',
        'authors',
        'indexed_year'
    ]]
    .sort_values(['author_query', 'indexed_year'])
)

# Tampilkan sebagai teks / tabel
# print(output_text.to_string(index=False))
output_text
```

## 2️⃣ 5 Dosen – Tren Produktivitas Paling Meningkat (5 Tahun)

```python
## 2️⃣ 5 Dosen – Tren Produktivitas Paling Meningkat (5 Tahun)
last5 = crossref[crossref['published_year'] >= year_1-4]
trend = last5.groupby(['author_query','published_year'])['doi'].count().unstack(fill_value=0)
trend['growth'] = trend.iloc[:,-1] - trend.iloc[:,0]


k2 = trend.sort_values('growth', ascending=False).head(5)


k2.iloc[:,:-1].T.plot(marker='o')
plt.title('Tren Produktivitas 5 Dosen Teratas (5 Tahun)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Publikasi')
plt.show()
```

## 3️⃣ Distribusi Kontribusi Publikasi per Fakultas (1 Tahun)

```python
## 3️⃣ Distribusi Kontribusi Publikasi per Fakultas (1 Tahun)

import pandas as pd
import matplotlib.pyplot as plt
import squarify
import numpy as np

# 3. Agregasi Data untuk Treemap
# Pastikan year_1 didefinisikan, misal: year_1 = 2023
k3 = (crossref[crossref['indexed_year'] == year_1]
      .dropna(subset=['programs_name'])
      .groupby('programs_name')['doi']
      .count()
      .sort_values(ascending=False))

# 4. Visualisasi Treemap
plt.figure(figsize=(14, 8))

# Membuat palet warna yang menarik
colors = plt.cm.viridis(np.linspace(0, 1, len(k3)))

# Format label: Nama Fakultas + Jumlah
labels = [f'{idx}\n({val})' for idx, val in zip(k3.index, k3.values)]

squarify.plot(
    sizes=k3.values, 
    label=labels, 
    alpha=0.7, 
    color=colors,
    edgecolor="white",
    linewidth=1.5
)

plt.title(f'Distribusi Kontribusi Publikasi Per Fakultas ({year_1})', fontsize=16, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.show()
```

## 4️⃣ Informasi rasio persentase penulis utama dan ko-author per dosen pada setiap fakultas dalam periode 1 tahun

```python
# coba author_query huruf pertama menjadi besar 
crossref['author_query'] = crossref['author_query'].str.title()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

nama_dosen = "IRAWAN AFRIANTO"
df_eddy = crossref[
    (crossref['author_query'].str.contains(nama_dosen, case=False, na=False))
]

df_eddy = df_eddy.copy()

def get_role(author_string, target_author):
    if pd.isna(author_string):
        return None
    
    authors = [a.strip() for a in author_string.split(';')]
    
    if authors[0].lower() == target_author.lower():
        return 'Penulis Utama'
    else:
        return 'Co-Author'

df_eddy['role'] = df_eddy['authors'].apply(
    lambda x: get_role(x, nama_dosen)
)


role_count = df_eddy['role'].value_counts()
utama = role_count.get('Penulis Utama', 0)
co = role_count.get('Co-Author', 0)

print(f"Rasio Penulis Utama : Co-Author = {utama} : {co}")



```

```python
# Tahun publikasi & indeks
crossref['indexed_year'] = pd.to_datetime(crossref['indexed_date_time'], errors='coerce').dt.year
crossref['published_year'] = pd.to_datetime(crossref['published'], errors='coerce').dt.year


# Jumlah penulis
crossref['author_count'] = crossref['authors'].fillna('').apply(lambda x: len(x.split(';')) if x else 0)


# Filter data valid
crossref = crossref[crossref['indexed_year'].notna()]
```

```python
# Tahun terbaru
year_1 = crossref['indexed_year'].max()

# Filter data 1 tahun
df_year = crossref[crossref['indexed_year'] == year_1]

# Hitung jumlah artikel dosen per jurnal per fakultas
journal_count = (
    df_year
    .groupby(['programs_name', 'title'])
    .size()
    .reset_index(name='jumlah_artikel')
)

# Ambil 5 jurnal teratas per fakultas
top_5_journal_per_faculty = (
    journal_count
    .sort_values('jumlah_artikel', ascending=False)
    .groupby('programs_name', group_keys=False)
    .head(5)
)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))

sns.barplot(
    data=top_5_journal_per_faculty,
    x='jumlah_artikel',
    y='title',
    hue='programs_name',
    dodge=False
)

plt.title(
    "5 Jurnal dengan Jumlah Artikel Dosen Terindeks Terbanyak\nper Fakultas (1 Tahun)",
    fontsize=14,
    weight='bold'
)

plt.xlabel("Jumlah Artikel Dosen")
plt.ylabel("Jurnal")

plt.legend(
    title="Fakultas",
    bbox_to_anchor=(1.02, 1),
    loc='upper left'
)

plt.tight_layout()
plt.show()
```
