```python
# ubah kolom container_title menjadi journal_title
df = df.rename(columns={"container_title": "journal_title"})

# ubah kolom short_container_title menjadi short_journal_title
df = df.rename(columns={"short_container_title": "short_journal_title"})

# hapus kolom all_links
df = df.drop(columns=['all_links'])

# hapus data kolom score
df = df.drop(columns=['score'])

# hapus data kolom abstract
df = df.drop(columns=['abstract'])

# Hapus data kolom source
df = df.drop(columns=['source'])
```

```python
# data apa saja yang ada di fullname tapi tidak ada di author_query
dfAuthor[~dfAuthor['fullname'].isin(df['author_query'])]
```

```python
# cari nama data nama ACHMAD DEPTIAN DJENUARI RIZKY di df
df[df['author_query'] == 'ACHMAD DEPTIAN DJENUARI RIZKY']
```

```python
# ambil id_sinta dari dfAuthor ke df berdasarkan author_query dan fullname
df = df.merge(dfAuthor[['fullname', 'id_sinta']], left_on='author_query', right_on='fullname', how='left')
```

```python
df
```

```python
# id_sinta jadi di awal kolom dan hapus author_query dan fullname
cols = df.columns.tolist()
cols = ['id_sinta'] + [col for col in cols if col not in ['id_sinta', 'author_query', 'fullname']]
df = df[cols]
df
```

```python
# buatkan id works unik mulai dari 1
df.insert(0, 'id_works', range(1, 1 + len(df)))
```
