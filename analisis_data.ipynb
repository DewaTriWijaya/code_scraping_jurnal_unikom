```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


plt.rcParams['figure.figsize'] = (10,6)
sns.set_style('whitegrid')
```

```python
dfAuthors = pd.read_csv('author/Data Done/Tabel Authors Final Unikom.csv')
dfAuthors.head()
```

```python
dfArticles = pd.read_csv('jurnal/crossref/Data Done/Tabel Article Final Unikom.csv')
dfArticles.head()
```

# 1 - Informasi 10 dosen dengan jumlah publikasi jurnal terindeks terbanyak dalam periode 1 tahun.

```python
dfArticles['indexed_year'] = pd.to_datetime(dfArticles['indexed_date_time'], errors='coerce').dt.year
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# tahun terbaru
year_1 = dfArticles['indexed_year'].max()

# hitung publikasi per dosen (id_sinta)
k1 = (
    dfArticles[dfArticles['indexed_year'] == year_1]
    .groupby('id_sinta')['doi']
    .count()
    .reset_index(name='jumlah_publikasi')
)

# ambil top 10
k1 = k1.sort_values('jumlah_publikasi', ascending=False).head(10)

# gabungkan dengan dfAuthors untuk ambil fullname
k1 = k1.merge(
    dfAuthors[['id_sinta', 'fullname']],
    on='id_sinta',
    how='left'
)

# ======================
# PLOT (HORIZONTAL BAR)
# ======================
plt.figure(figsize=(10, 6))

sns.barplot(
    data=k1,
    x='jumlah_publikasi',
    y='fullname',
    orient='h'
)

plt.title(f'Top 10 Dosen dengan Publikasi Terbanyak ({year_1})')
plt.xlabel('Jumlah Publikasi')
plt.ylabel('Dosen')
plt.tight_layout()
plt.show()
```

```python
# tolong dari ke 10 di atas berikan detail authornya
top_authors = k1['id_sinta'].tolist()
detailed_info = dfAuthors[dfAuthors['id_sinta'].isin(top_authors)]
detailed_info = detailed_info.merge(k1[['id_sinta', 'jumlah_publikasi']], on='id_sinta', how='left')
print(detailed_info[['id_sinta', 'fullname', 'jumlah_publikasi', 'major', 'nidn']])
```

# 2 - Informasi 5 dosen dengan fluktuasi publikasi dalam periode 5 tahun terakhir.

```python
dfArticles['published_year'] = pd.to_datetime(dfArticles['published'], errors='coerce').dt.year
```

```python
# data 5 tahun terakhir
last5 = dfArticles[dfArticles['published_year'] >= year_1 - 4]

trend = (
    last5
    .groupby(['id_sinta', 'published_year'])['doi']
    .count()
    .unstack(fill_value=0)
)

# hitung growth (tahun terakhir - tahun pertama)
trend['growth'] = trend.iloc[:, -1] - trend.iloc[:, 0]

# ambil 5 dosen dengan growth tertinggi
k2 = trend.sort_values('growth', ascending=False).head(5)

# ======================
# GANTI id_sinta â†’ fullname
# ======================
k2 = k2.merge(
    dfAuthors[['id_sinta', 'fullname']],
    left_index=True,
    right_on='id_sinta',
    how='left'
)

k2 = k2.set_index('fullname')

# ======================
# PLOT
# ======================
k2.iloc[:, :-1].T.plot(marker='o', figsize=(10, 6))

plt.title('Tren Produktivitas 5 Dosen Teratas (5 Tahun)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Publikasi')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

# 3 - Informasi distribusi publikasi setiap fakultas terhadap total publikasi UNIKOM dalam periode 1 tahun

```python
faculty_map = (
    dfAuthors
    .drop_duplicates('id_sinta')
    .set_index('id_sinta')['faculty']
)

dfArticles['faculty'] = dfArticles['id_sinta'].map(faculty_map)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify
import numpy as np

# 3. Agregasi Data untuk Treemap
# Pastikan year_1 didefinisikan, misal: year_1 = 2023
k3 = (dfArticles[dfArticles['indexed_year'] == year_1]
      .dropna(subset=['faculty'])
      .groupby('faculty')['doi']
      .count()
      .sort_values(ascending=False))

# 4. Visualisasi Treemap
plt.figure(figsize=(14, 8))

# Membuat palet warna yang menarik
colors = plt.cm.viridis(np.linspace(0, 1, len(k3)))

# Format label: Nama Fakultas + Jumlah
labels = [f'{idx}\n({val})' for idx, val in zip(k3.index, k3.values)]

squarify.plot(
    sizes=k3.values, 
    label=labels, 
    alpha=0.7, 
    color=colors,
    edgecolor="white",
    linewidth=1.5
)

plt.title(f'Distribusi Kontribusi Publikasi Per Fakultas ({year_1})', fontsize=16, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.show()
```

# 4 - Informasi rasio persentase penulis dan ko author per dosen pada setiap fakultas dalam periode 1 tahun.

```python
dfEmpat = dfArticles.merge(
    dfAuthors[['id_sinta', 'fullname']],
    on='id_sinta',
    how='left'
)
```

```python
import pandas as pd

# ======================
# LIST DOSEN
# ======================
list_dosen = dfAuthors["fullname"].tolist()

# ======================
# FUNGSI ROLE
# ======================
def get_role(author_string, target_author):
    if pd.isna(author_string):
        return None
    
    authors = [a.strip() for a in author_string.split(';')]
    
    if authors and authors[0].lower() == target_author.lower():
        return 'Penulis Utama'
    else:
        return 'Co-Author'


# ======================
# HITUNG RASIO PER DOSEN
# ======================
result = []

for dosen in list_dosen:
    df_dosen = dfEmpat[
        dfEmpat['fullname'].str.contains(dosen, case=False, na=False)
    ].copy()

    if df_dosen.empty:
        result.append({
            'fullname': dosen,
            'penulis_utama': 0,
            'co_author': 0,
            'rasio_utama_co': '0 : 0'
        })
        continue

    df_dosen['role'] = df_dosen['authors'].apply(
        lambda x: get_role(x, dosen)
    )

    role_count = df_dosen['role'].value_counts()

    utama = role_count.get('Penulis Utama', 0)
    co = role_count.get('Co-Author', 0)

    result.append({
        'fullname': dosen,
        'penulis_utama': utama,
        'co_author': co,
        'rasio_utama_co': f"{utama} : {co}"
    })

# ======================
# DATAFRAME HASIL
# ======================
df_rasio_dosen = pd.DataFrame(result)

df_rasio_dosen
```

# 5 - Informasi rasio persentase penulis dan ko author per dosen pada setiap fakultas dalam periode 1 tahun.

```python
k6 = (
    dfArticles[
        (dfArticles['indexed_year'] >= year_1 - 4) &
        (dfArticles['type'] != 'journal-article')
    ]
    .groupby(['indexed_year', 'type'])['doi']
    .count()
    .unstack(fill_value=0)
)

# Bar chart horizontal
for year in k6.index:
    k6.loc[year].sort_values().plot(
        kind='barh',
        figsize=(6, 4),
        title=f'Tipe Publikasi Tahun {year}'
    )
    plt.xlabel('Jumlah Publikasi')
    plt.tight_layout()
    plt.show()

```

# 6 - Informasi 5 Dosen dengan Dampak Penelitian Tertinggi per Fakultas berdasarkan Analisis Komposit Metrik Kutipan (h-index, g-index, dan Total Kutipan)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# hitung total publikasi per fakultas
pub_faculty = (
    dfArticles
    .groupby('faculty')['url']
    .count()
    .sort_values(ascending=True)
)

plt.figure(figsize=(8,5))
sns.barplot(x=pub_faculty.values, y=pub_faculty.index)
plt.title('Total Publikasi per Fakultas')
plt.xlabel('Jumlah Publikasi')
plt.ylabel('Fakultas')
plt.show()
```

```python
year_max = dfArticles['indexed_year'].max()

last5 = dfArticles[dfArticles['indexed_year'] >= year_max - 4]

trend_faculty = (
    last5
    .groupby(['indexed_year','faculty'])['url']
    .count()
    .unstack(fill_value=0)
)

plt.figure(figsize=(10,5))
trend_faculty.plot(marker='o')
plt.title('Tren Publikasi Fakultas (5 Tahun Terakhir)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Publikasi')
plt.legend(title='Fakultas', bbox_to_anchor=(1.05,1))
plt.show()
```

```python
faculty_year = (
    dfArticles
    .groupby(['faculty','indexed_year'])['url']
    .count()
    .reset_index()
)

ipf = (
    faculty_year
    .groupby('faculty')['url']
    .mean()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))
sns.barplot(x=ipf.values, y=ipf.index)
plt.title('Indeks Produktivitas Fakultas')
plt.xlabel('Rata-rata Publikasi per Tahun')
plt.ylabel('Fakultas')
plt.show()
```
