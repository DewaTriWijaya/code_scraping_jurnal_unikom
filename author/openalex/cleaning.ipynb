```python
import pandas as pd

authors = pd.read_csv('authors.csv')

# ubah column id menjadi id_author
authors.rename(columns={'id': 'id_author'}, inplace=True)

# hapus id_sinta, country, affiliation_name
authors.drop(columns=['id_sinta', 'country', 'affiliation_name'], inplace=True)
```

```python
import pandas as pd

df = pd.read_csv('dosen_openalex_by_name_only.csv')

# ubah = dan " menjadi kosong 
df.replace({'=': '', '"': ''}, regex=True, inplace=True)

df
```

```python
# cek open_alex id yang kosong count
df['openalex_id'].isnull().sum()
```

```python
# cek open_alex id yang kosong
df[df['openalex_id'].isnull()]
```

```python
# id_author bernama Susmini, suryatno, annisa, nang primadi, mochhamad valuenya pindahkan ke nama lalu id_author diisi dengan kosong
df.loc[df['id_author'].isin(['SUSMINI INDRIANI L', 'SURYATNO W SOEGOTO', 'ANNISA SUCI RAMADANI', 'NANG PRIMADI TABRANI', 'MOCHAMAD EKKI HARDIAWA']), 'nama'] = df['id_author']
df.loc[df['id_author'].isin(['SUSMINI INDRIANI L', 'SURYATNO W SOEGOTO', 'ANNISA SUCI RAMADANI', 'NANG PRIMADI TABRANI', 'MOCHAMAD EKKI HARDIAWA']), 'id_author'] = ''
```

```python
import numpy as np
# ubah empty string menjadi NaN (numpy NaN)
df['id_author'] = df['id_author'].replace('', np.nan)

# ubah kolom menjadi float agar bisa ditambahkan
df['id_author'] = pd.to_numeric(df['id_author'], errors='coerce')

# isi NaN dengan nilai sebelumnya + 1
for i in range(len(df)):
    if pd.isna(df.loc[i, 'id_author']):
        df.loc[i, 'id_author'] = df.loc[i-1, 'id_author'] + 1

# ubah kembali ke string jika diperlukan
df['id_author'] = df['id_author'].astype(int).astype(str)



# cek open_alex id yang kosong
df[df['openalex_id'].isnull()]
```

```python
# tambah nidn dari authors_cleaned.csv berdasarkan nama yang sama
df = df.merge(authors[['fullname', 'nidn', 'programs_name', 'gelar_depan', 'gelar_belakang']], left_on='nama', right_on='fullname', how='left')
```

```python
# save to csv
df.to_csv('dosen_openalex_by_name_only_cleaned.csv', index=False)
```
